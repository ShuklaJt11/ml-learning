# K Means Clustering

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8, random_state=101)
```

```python
plt.figure(figsize=(10, 7))
sns.scatterplot(data[0][:, 0], data[0][:, 1], hue=data[1], palette='tab10')
```

```python
from sklearn.cluster import KMeans
```

```python
kmm = KMeans(n_clusters=4)
```

```python
kmm.fit(data[0])
```

```python
kmm.cluster_centers_
```

```python
kmm.labels_
```

```python
plt.style.use('seaborn')
fig, (a1, a2) = plt.subplots(1, 2, figsize=(15, 7), sharey=True)
a1.set_title('K Means')
a1.scatter(data[0][:, 0], data[0][:, 1], c=kmm.labels_, cmap='rainbow')
a2.set_title('Original')
a2.scatter(data[0][:, 0], data[0][:, 1], c=data[1], cmap='rainbow')
```
