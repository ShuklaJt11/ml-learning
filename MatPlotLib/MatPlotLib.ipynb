# MatplotLib

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

### Functional Method

```python
plt.plot(x, y)
```

```python
plt.plot(x, y)
plt.xlabel('X Value')
plt.ylabel('X Sqaured')
plt.title('Square Graph')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')
plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

### Object Oriented Method

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y)
axes.set_xlabel('X Value')
axes.set_ylabel('X Squared')
axes.set_title('Squared Graph')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
axes1.plot(x, y)
axes2.plot(y, x)
axes2.set_title('Small One')
axes1.set_title('Big One')
axes1.set_xlabel('X')
axes1.set_ylabel('Y = X^2')
axes2.set_xlabel('Y')
axes2.set_ylabel('X')
```

```python
fig, axes = plt.subplots()
axes.plot(x, y)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
for current_axis in axes:
    current_axis.plot(x, y)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y)
axes[1].plot(y, x)
```

### Figure Size and DPI

```python
fig = plt.figure(figsize=(3, 2), dpi=100)
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
```

```python
fig = plt.figure(figsize=(3, 2))
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
```

```python
fig = plt.figure(figsize=(8, 2), dpi=100)
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
```

```python
fig, axes = plt.subplots(figsize=(3, 2), dpi=100)
axes.plot(x, y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(5, 2), dpi=100)
axes[0].plot(x, y)
axes[1].plot(y, x)
plt.tight_layout()
```

```python
fig = plt.figure(figsize=(5, 4), dpi=100)
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x ** 2)
ax.plot(x, x ** 3)
```

```python
fig = plt.figure(figsize=(5, 4), dpi=100)
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x ** 2, label='x ^ 2')
ax.plot(x, x ** 3, label='x ^ 3')
ax.legend()
```

```python
fig = plt.figure(figsize=(5, 4), dpi=100)
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x ** 2, label='x ^ 2')
ax.plot(x, x ** 3, label='x ^ 3')
ax.legend(loc=10)
```

```python
fig = plt.figure(figsize=(5, 4), dpi=100)
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x ** 2, label='x ^ 2')
ax.plot(x, x ** 3, label='x ^ 3')
ax.legend(loc=(0.1, 0.1))
```

### Plot Appearance

**Plot and Marker**

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='yellow') #Color string
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='#FF8C00')  #RGB Hex Code
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='purple', linewidth=2)
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='purple', linewidth=3, alpha=0.5)
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='purple', lw=3)
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='purple', lw=2, linestyle='--')
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='purple', lw=2, ls='-.')
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='purple', lw=2, marker='o')
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='purple', lw=1, marker='*')
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='purple', lw=2, marker='+', markersize=20)
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='purple', lw=2, marker='o', markersize=20, markerfacecolor='yellow', markeredgewidth=3, markeredgecolor='red')
```

**Axis Control**

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='purple', lw=2, ls='--')
ax.set_xlim([0, 1])
ax.set_ylim([0, 2])
```
