# NumPy Arrays

```python
import numpy as np
```

```python
lst = [1, 2, 3, 4, 5]
```

```python
np.array(lst)
```

#### np.arange() is similar to range() function

```python
np.arange(0, 20, 3)
```

#### To get array of zeros 

```python
np.zeros(5)
```

```python
np.zeros((5, 5))
```

```python
np.zeros((3, 4))
```

#### To get array of ones

```python
np.ones(3)
```

```python
np.ones((2, 4))
```

#### np.linspace(start, end, num) will provide 'num' diffrent points between 'start' (inclusive) and 'end' (inclusive) that are equidistant to each other

```python
np.linspace(0, 20, 10)
```

```python
np.linspace(0, 5, 100)
```

#### Identity Matrix

```python
np.eye(5)
```

### Random Number Generation

#### Uniform Distribution from 0 to 1

```python
np.random.rand(50)
```

```python
np.random.rand(3, 20)
```

#### Normal / Gaussian Distribution centered around 0

```python
np.random.randn(30)
```

```python
np.random.randn(2, 12)
```

#### Random integers from 'low' (inclusive) to 'high' (exclusive)

```python
np.random.randint(1, 100)
```

```python
np.random.randint(1, 100, 25)
```
