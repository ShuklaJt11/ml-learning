# NumPy Indexing and Selection

```python
import numpy as np
```

## 1D Array

```python
arr = np.arange(0, 11)
```

```python
arr
```

```python
arr[8]
```

```python
arr[1:6]
```

### Broadcasting

```python
arr[1:6] = 100
```

```python
arr
```

```python
arr = np.arange(0, 11)
```

```python
arr
```

__Note: When a slice is asigned to variable, data is not copied but is just a view of original array. This is done to preserve memory.__

```python
slice_arr = arr[5:9]
```

```python
slice_arr
```

```python
slice_arr[:] = 99
```

```python
slice_arr
```

```python
arr
```

__If you specifically require a copy of original array, use copy() method__

```python
arr = np.arange(0, 11)
```

```python
arr
```

```python
arr_cp = arr.copy()
```

```python
arr_cp
```

```python
arr_cp[:] = 100
```

```python
arr_cp
```

```python
arr
```

## 2D Array (Matrix)

```python
arr_2d = np.array([
    [5, 10, 15],
    [20, 25, 30],
    [35, 40, 45]
])
```

```python
arr_2d
```

```python
arr_2d[1][1]
```

```python
arr_2d[1, 1]
```

```python
arr_2d[1, ]
```

```python
arr_2d[1, 0:2]
```

```python
arr_2d[:2, 1:]
```

```python
arr_2d[:, 1]
```

## Conditional Selection

```python
arr = np.arange(1, 11)
```

```python
arr
```

```python
arr > 6
```

```python
bool_arr = arr < 4
```

```python
bool_arr
```

```python
arr[bool_arr]
```

```python
arr[arr > 8]
```

```python
arr[arr % 2 == 0]
```

```python
arr[arr % 2 != 0]
```

```python
arr_2d = np.arange(50).reshape(5, 10)
```

```python
arr_2d
```

```python
arr_2d[arr_2d % 2 == 0]
```
