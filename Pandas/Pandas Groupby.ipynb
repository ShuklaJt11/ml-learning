# Pandas Groupby

```python
import numpy as np
import pandas as pd
```

```python
data = {'Company':['GOOG'] * 2 + ['MSFT'] * 2 + ['FB'] * 2,
       'Person': ['Sam', 'Dean', 'Castiel', 'Crowly', 'Chuck', 'Jack'],
       'Sales': [200, 143, 340, 124, 243, 350]}
```

```python
data
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df.groupby('Company')
```

```python
by_comp = df.groupby('Company')
```

```python
by_comp.mean()
```

```python
by_comp.sum()
```

```python
by_comp.std()
```

```python
by_comp.sum().loc['FB']
```

```python
df.groupby('Company').std().loc['MSFT']
```

```python
df.groupby('Company').count()
```

```python
df.groupby('Company').max()
```

```python
df.groupby('Company').min()
```

```python
df.groupby('Company').describe()
```

```python
df.groupby('Company').describe().transpose()
```
