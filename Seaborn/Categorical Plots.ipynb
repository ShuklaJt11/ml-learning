# Categorical Plots

```python
import numpy as np
import pandas as pd
```

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

### Bar Plot

```python
sns.barplot(x='sex', y='total_bill', data=tips)
```

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

*Note: Default estimator value is 'Arithematic Mean'*

### Count Plot

```python
sns.countplot(x='sex', data=tips)
```

*Note: It is basically just a bar plot with estimator fuction as count*

### Box Plot and Violin Plot

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

### Strip Plot

```python
sns.stripplot(x='day', y='tip', data=tips)
```

```python
sns.stripplot(x='day', y='tip', data=tips, jitter=False)
```

```python
sns.stripplot(x='day', y='tip', data=tips, hue='sex', dodge=True)
```

```python
sns.stripplot(x='day', y='tip', data=tips, hue='sex')
```

### Swarm Plot

```python
sns.swarmplot(x='day', y='tip', data=tips)
```

```python
sns.violinplot(x='day', y='tip', data=tips)
sns.swarmplot(x='day', y='tip', data=tips, color='black')
```

### Factor Plot

```python
sns.factorplot(x='day', y='tip', data=tips, kind='bar')
```

```python
sns.factorplot(x='day', y='tip', data=tips, kind='strip')
```
