# Matrix Plots

```python
import numpy as np
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
```

```python
tips.head(2)
```

```python
flights.head()
```

*Note: Data should be in matrix form. So converting tables into matrices in below 2 cells*

```python
tc = tips.corr()
```

```python
fp = flights.pivot_table(index='month', columns='year', values='passengers')
```

### Heat Map

```python
sns.heatmap(tc)
```

```python
sns.heatmap(tc, annot=True)
```

```python
sns.heatmap(tc, annot=True, cmap='coolwarm')
```

```python
sns.heatmap(fp)
```

```python
sns.heatmap(fp, cmap='coolwarm')
```

```python
sns.heatmap(fp, linecolor='white', linewidth=2)
```

```python
sns.heatmap(fp, linecolor='white', linewidth=0.5)
```

### Cluster Map

```python
sns.clustermap(fp)
```

```python
sns.clustermap(fp, cmap='coolwarm', standard_scale=1)
```
