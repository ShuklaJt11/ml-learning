# Regression Plots

```python
import numpy as np
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

### LM Plot

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='time')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='time', markers=['o', 'v'])
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='time', markers=['o', 'v'],
           scatter_kws={
               's': 100
           })
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='time')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', row='smoker')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', row='smoker', hue='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', hue='sex', aspect=.6, height=8)
```
